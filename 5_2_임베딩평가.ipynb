```python
import numpy as np
from konlpy.tag import Mecab
import scipy.stats as st
import pdb
tokenizer = Mecab()



# token vector들을 lookup한 뒤 평균을 취한다
def get_word_vector(model_type, model, word):
    vector = []
    try:
      if model_type == 'fasttext' or model_type == 'w2v':
        vector = model.wv[word]
      elif model_type == 'glove':
        vector = glove.word_vectors[glove.dictionary[word]]
    except:
      vector = np.zeros(100) # dim is 100
    return vector

def get_sentence_vector(model_type, model,  sentence):
    tokens = tokenizer.morphs(sentence)
    token_vecs = []
    for token in tokens:
        token_vecs.append(get_word_vector(model_type, model, token))
    return np.mean(token_vecs, axis=0)
  
def word_sim_test(model_type, model, test_fname):
    actual_sim_list, pred_sim_list = [], []
    missed = 0
    with open(test_fname, 'r') as pairs:
        for pair in pairs:

            w1, w2, actual_sim = pair.strip().split(",")
            try:
              w1_vec = get_sentence_vector(model_type,model,w1)
              w2_vec = get_sentence_vector(model_type,model, w2)
              score = np.dot(w1_vec, w2_vec)
              actual_sim_list.append(float(actual_sim))
              pred_sim_list.append(score)
            except KeyError:
                missed += 1
    spearman, _ = st.spearmanr(actual_sim_list, pred_sim_list)
    pearson, _ = st.pearsonr(actual_sim_list, pred_sim_list)
    print("spearman corr:", spearman, ", pearson corr:", pearson, ", # of errors:", missed)

test_file_path = root_path + '/data/test_file/kor_ws353.csv'

print("w2v")
word_sim_test('w2v',w2v,test_file_path )
print("fasttext")
word_sim_test('fasttext',fasttext,test_file_path )
print("glove")
word_sim_test('glove',glove,test_file_path )

```

## 단어 유추 평가

갑-을+병 = 정 의 관계를 파악합니다
 

```python
from sklearn.preprocessing import normalize
def most_similar_by_vector(model_type, model, query_vec, topn=10):
    vecs= []
    if model_type == 'fasttext' or model_type == 'w2v':
        vecs = model.wv.vectors
        words = model.wv.index2word
    elif model_type == 'glove':
        vecs = model.word_vectors
        words = model.dictionary
    unit_vecs = normalize(vecs, norm='l2', axis=1)

    query_vec_norm = np.linalg.norm(query_vec)
    if query_vec_norm != 0:
        query_unit_vec = query_vec / query_vec_norm
    else:
        query_unit_vec = query_vec
    scores = np.dot(unit_vecs, query_unit_vec)
    topn_candidates = sorted(zip(words, scores), key=lambda x: x[1], reverse=True)[1:topn+1]
    return topn_candidates


def get_analogy_vector(model_type, model, words):
    if len(words) == 3:
        token_1 = get_sentence_vector(model_type, model, words[0])
        token_2 = get_sentence_vector(model_type, model, words[1])
        token_3 = get_sentence_vector(model_type, model, words[2])
        result = token_1 - token_2 + token_3
    else:
        result = np.zeros(100)
    return result


def word_analogy_test(model_type, model,  test_fname, topn=30, verbose=False):
    correct, total, missed = 0, 0, 0
    with open(test_fname, 'r', encoding='utf-8') as f:
        for line in f:
            if line.startswith("#") or len(line) <= 1:
                continue
            words = line.strip().split(" ")
            query_vecs = get_analogy_vector(model_type, model, words[:-1])
            # try:
            word_with_scores = most_similar_by_vector(model_type, model, query_vecs, topn)
            if verbose:
                print(words[0] + " - " + words[1] + " + " + words[2])
                print("correct answer:", words[3])
                print("predicted answers:", word_with_scores)
                print("")
            similar_words = [el[0] for el in word_with_scores]
            if words[-1] in similar_words:
                correct += 1
            # except:
                # missed += 1
            total += 1
    print("# of correct answer:", correct, ", # of data:", total, ", # of errors:", missed)



test_file_path = root_path + '/data/test_file/kor_analogy_semantic.txt'
word_analogy_test('w2v',w2v,test_file_path  )
word_analogy_test('fasttext',fasttext,test_file_path )
word_analogy_test('glove',glove,test_file_path )
```

## 단어 임베딩 시각화

```python
import numpy as np
import pandas as pd
import networkx as nx
from sklearn.manifold import TSNE
from sklearn.metrics.pairwise import cosine_similarity

from bokeh.io import export_png, output_notebook, show
from bokeh.plotting import figure
from bokeh.models import Plot, Range1d, MultiLine, Circle, HoverTool, TapTool, BoxSelectTool, LinearColorMapper, ColumnDataSource, LabelSet, SaveTool, ColorBar, BasicTicker
from bokeh.models.graphs import from_networkx, NodesAndLinkedEdges, EdgesAndLinkedNodes
from bokeh.palettes import Spectral8



def _visualize_words(words, vecs, palette="Viridis256", filename=root_path + '/image/image.png',
                    use_notebook=True):
    tsne = TSNE(n_components=2)
    tsne_results = tsne.fit_transform(vecs)
    df = pd.DataFrame(columns=['x', 'y', 'word'])
    df['x'], df['y'], df['word'] = tsne_results[:, 0], tsne_results[:, 1], list(words)
    source = ColumnDataSource(ColumnDataSource.from_df(df))
    labels = LabelSet(x="x", y="y", text="word", y_offset=8,
                      text_font_size="15pt", text_color="#555555",
                      source=source, text_align='center')
    color_mapper = LinearColorMapper(palette=palette, low=min(tsne_results[:, 1]), high=max(tsne_results[:, 1]))
    plot = figure(plot_width=900, plot_height=900)
    plot.scatter("x", "y", size=12, source=source, color={'field': 'y', 'transform': color_mapper}, line_color=None,
                 fill_alpha=0.8)
    plot.add_layout(labels)
    if use_notebook:
        output_notebook()
        show(plot)
    else:
        export_png(plot, filename)
        print("save @ " + filename)
```

```python

test_file_path = root_path + '/data/test_file/kor_analogy_semantic.txt'
def visualize_words( model_type, model,words_fname, palette="Viridis256"):
    words = set()
    for line in open(words_fname, 'r', encoding='utf-8'):
        if not line.startswith("#"):
            for word in line.strip().split(" "):
                if len(word) > 0:
                    words.add(word)
    vecs = np.array([get_sentence_vector(model_type, model, word) for word in words])
    _visualize_words(words, vecs, palette)
```

```python
visualize_words('w2v', w2v, test_file_path)
```

```python
visualize_words('fasttext', fasttext, test_file_path)
```

```python
visualize_words('glove', glove, test_file_path)
```

```python
# 단어간 벡터 유사도 시각화
def _visualize_between_words(words, vecs, palette="Viridis256", filename=root_path + '/image/between-words.png',
                            use_notebook=True):
    df_list = []
    for word1_idx, word1 in enumerate(words):
        for word2_idx, word2 in enumerate(words):
            vec1 = vecs[word1_idx]
            vec2 = vecs[word2_idx]
            if np.any(vec1) and np.any(vec2):
                score = cosine_similarity(X=[vec1], Y=[vec2])
                df_list.append({'x': word1, 'y': word2, 'similarity': score[0][0]})
    df = pd.DataFrame(df_list)
    color_mapper = LinearColorMapper(palette=palette, low=1, high=0)
    TOOLS = "hover,save,pan,box_zoom,reset,wheel_zoom"
    p = figure(x_range=list(words), y_range=list(reversed(list(words))),
               x_axis_location="above", plot_width=900, plot_height=900,
               toolbar_location='below', tools=TOOLS,
               tooltips=[('words', '@x @y'), ('similarity', '@similarity')])
    p.grid.grid_line_color = None
    p.axis.axis_line_color = None
    p.axis.major_tick_line_color = None
    p.axis.major_label_standoff = 0
    p.xaxis.major_label_orientation = 3.14 / 3
    p.rect(x="x", y="y", width=1, height=1,
           source=df,
           fill_color={'field': 'similarity', 'transform': color_mapper},
           line_color=None)
    color_bar = ColorBar(ticker=BasicTicker(desired_num_ticks=5),
                         color_mapper=color_mapper, major_label_text_font_size="7pt",
                         label_standoff=6, border_line_color=None, location=(0, 0))
    p.add_layout(color_bar, 'right')
    if use_notebook:
        output_notebook()
        show(p)
    else:
        export_png(p, filename)
        print("save @ " + filename)

def visualize_between_words(model_type, model, words_fname, palette="Viridis256"):
    words = set()
    for line in open(words_fname, 'r'):
        if not line.startswith("#"):
            for word in line.strip().split(" "):
                if len(word) > 0:
                    words.add(word)
    vecs = [get_sentence_vector(model_type, model, word) for word in words]
    _visualize_between_words(words, vecs, palette)

```

```python
visualize_between_words('fasttext', fasttext, test_file_path)
```
